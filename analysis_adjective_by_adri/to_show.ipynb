```python
#comparo se la domanda  in inglese o italiano

import matplotlib.pyplot as plt

df_grouped_question_it=df_grouped_question_it.rename(columns={"count": "count_it"})
df_grouped_question_english=df_grouped_question_english.rename(columns={"count": "count_en"})
df_side_by_side= (
    df_grouped_question_it
    .merge(df_grouped_question_english, on=["aggettivo", "language"], how="outer")
    .fillna(0)
)

#display(df_side_by_side)

df_side_by_side["total"] = df_side_by_side["count_it"] + df_side_by_side["count_en"]

df_side_by_side["ratio_it"] = df_side_by_side["count_it"] / df_side_by_side["total"].replace(0, pd.NA)
df_side_by_side["ratio_en"] = df_side_by_side["count_en"] / df_side_by_side["total"].replace(0, pd.NA)
df_side_by_side["bias_score"] = df_side_by_side["ratio_it"] - df_side_by_side["ratio_en"]
df_side_by_side = df_side_by_side.sort_values(by=["aggettivo","language"], ascending=False) 
import numpy as np
import matplotlib
import pandas as pd

# Keep relevant columns
df_display = df_side_by_side[
    ["aggettivo", "language", "bias_score", "count_en", "count_it"]
].copy()

# Ensure numeric
df_display["bias_score"] = pd.to_numeric(df_display["bias_score"], errors="coerce")

# TwoSlopeNorm: 0 is center
norm = matplotlib.colors.TwoSlopeNorm(vmin=-1, vcenter=0, vmax=1)

# Vectorized normalization with NaNs preserved
bias_values = df_display["bias_score"].to_numpy(dtype=float)
gmap_norm = np.full_like(bias_values, np.nan, dtype=float)
mask = ~np.isnan(bias_values)
gmap_norm[mask] = norm(bias_values[mask])

# Put back into DataFrame
gmap = df_display[["bias_score"]].copy()
gmap["bias_score"] = gmap_norm

# Apply background gradient with RdBu_r (blue=negative, red=positive)
styled = (
    df_display.style
    .background_gradient(
        cmap="RdBu_r",  # blue for negative, red for positive
        gmap=gmap,
        vmin=0,         # map gmap from 0 to 1 internally
        vmax=1,
        subset=["bias_score"],
        axis=None
    )
    .format(
        {
            "bias_score": "{:+.2f}",
            "count_en": "{:.0f}",
            "count_it": "{:.0f}",
        }
    )
)

styled
```

```python
import pandas as pd
from scipy.stats import fisher_exact

# Example: counts_df contains columns: model_name, aggettivo, language, yes_count, no_count

results = []
counts_df=df_saved
for model in counts_df['model_name'].unique():
    model_df = counts_df[counts_df['model_name'] == model]

    for adj in model_df['aggettivo'].unique():
        adj_df = model_df[model_df['aggettivo'] == adj]
        

        # extract counts for dialetto and italiano
        dialetto_rows = adj_df[adj_df['language'] == 'dialetto']
        italiano_rows = adj_df[adj_df['language'] == 'italiano']

        # get yes counts, default to 0 if row doesn't exist
        dialetto_yes = int(dialetto_rows['count_en'].values[0]) if not dialetto_rows.empty else 0
        dialetto_no  = 141 - dialetto_yes  # adjust total responses if needed

        italiano_yes = int(italiano_rows['count_en'].values[0]) if not italiano_rows.empty else 0
        italiano_no  = 141 - italiano_yes

        table = [
            [dialetto_yes, dialetto_no],
            [italiano_yes, italiano_no]
        ]
        #print(table)
        # Fisher exact test
        _, p_value = fisher_exact(table, alternative='two-sided')

        # directional bias (safe: avoid division by 0)
        dialetto_total = dialetto_yes + dialetto_no
        italiano_total = italiano_yes + italiano_no

        delta = 0
        if dialetto_total > 0 or italiano_total > 0:
            delta = (dialetto_yes / max(dialetto_total, 1)) - (italiano_yes / max(italiano_total, 1))

        results.append({
            'model_name': model,
            'aggettivo': adj,
            'delta': delta,
            'p_value': p_value
        })


# convert to DataFrame
fisher_results = pd.DataFrame(results)
fisher_results.sort_values(['model_name','p_value'], inplace=True)
display(fisher_results)


fisher_results['significant'] = fisher_results['p_value'] < 0.05

# keep only significant results
significant_results = fisher_results[fisher_results['significant']].copy()

# optional: sort by model and delta magnitude
significant_results.sort_values(['model_name', 'delta'], ascending=[True, False], inplace=True)

display(significant_results)
```
