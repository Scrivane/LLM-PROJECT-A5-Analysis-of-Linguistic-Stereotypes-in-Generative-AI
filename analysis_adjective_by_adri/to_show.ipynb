```python
import pandas as pd

from itertools import product


# Show full text in columns and all rows
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)

# Load JSONL

path_domanda_italian="../test_multiple_napoli/2from_dataset_answers_napoletano_templated.jsonl"

df = pd.read_json(path_domanda_italian, lines=True)


df['response'] = df['response_rephrased'].fillna(df['response'])
df["aggettivo"] = (
    df["prompt"]
    .str.strip()
    .str.rstrip(".")
    .str.split()
    .str[-1]
)

filtered_agg=df[df['aggettivo'].isin(['testarda'])]
filtered_agg=filtered_agg[df['language'].isin(['siciliano','dialetto'])]
aggettivi = df['aggettivo'].unique()
languages = df['language'].unique()
all_combinations = pd.DataFrame(list(product(aggettivi, languages)), columns=['aggettivo', 'language'])
df_yes_si = df[
    df["response"]
    .str.strip()
    .str.rstrip("!.,")
    .str.casefold()
    .isin(["yes", "si"])
]
counts = df_yes_si.groupby(['aggettivo', 'language']).size().reset_index(name='count')
df_grouped = all_combinations.merge(counts, on=['aggettivo', 'language'], how='left').fillna(0)
df_grouped['count'] = df_grouped['count'].astype(int)
df_grouped = df_grouped.sort_values(['aggettivo', 'language'], ascending=[False, False])
counts = filtered_agg['language'].value_counts()
print(counts)
display(df_grouped)




```

```python
#comparo se la domanda  in inglese o italiano

import matplotlib.pyplot as plt

df_grouped_question_it=df_grouped_question_it.rename(columns={"count": "count_it"})
df_grouped_question_english=df_grouped_question_english.rename(columns={"count": "count_en"})
df_side_by_side= (
    df_grouped_question_it
    .merge(df_grouped_question_english, on=["aggettivo", "language"], how="outer")
    .fillna(0)
)

#display(df_side_by_side)

df_side_by_side["total"] = df_side_by_side["count_it"] + df_side_by_side["count_en"]

df_side_by_side["ratio_it"] = df_side_by_side["count_it"] / df_side_by_side["total"].replace(0, pd.NA)
df_side_by_side["ratio_en"] = df_side_by_side["count_en"] / df_side_by_side["total"].replace(0, pd.NA)
df_side_by_side["bias_score"] = df_side_by_side["ratio_it"] - df_side_by_side["ratio_en"]
df_side_by_side = df_side_by_side.sort_values(by=["aggettivo","language"], ascending=False) 
import numpy as np
import matplotlib
import pandas as pd

# Keep relevant columns
df_display = df_side_by_side[
    ["aggettivo", "language", "bias_score", "count_en", "count_it"]
].copy()

# Ensure numeric
df_display["bias_score"] = pd.to_numeric(df_display["bias_score"], errors="coerce")

# TwoSlopeNorm: 0 is center
norm = matplotlib.colors.TwoSlopeNorm(vmin=-1, vcenter=0, vmax=1)

# Vectorized normalization with NaNs preserved
bias_values = df_display["bias_score"].to_numpy(dtype=float)
gmap_norm = np.full_like(bias_values, np.nan, dtype=float)
mask = ~np.isnan(bias_values)
gmap_norm[mask] = norm(bias_values[mask])

# Put back into DataFrame
gmap = df_display[["bias_score"]].copy()
gmap["bias_score"] = gmap_norm

# Apply background gradient with RdBu_r (blue=negative, red=positive)
styled = (
    df_display.style
    .background_gradient(
        cmap="RdBu_r",  # blue for negative, red for positive
        gmap=gmap,
        vmin=0,         # map gmap from 0 to 1 internally
        vmax=1,
        subset=["bias_score"],
        axis=None
    )
    .format(
        {
            "bias_score": "{:+.2f}",
            "count_en": "{:.0f}",
            "count_it": "{:.0f}",
        }
    )
)

styled
```

```python

heatmap_df = df_grouped_question_it.pivot_table(
    index="aggettivo",
    columns="language",
    values="count_it",
    fill_value=0
)

#display(heatmap_df)



df_bias = heatmap_df.copy()

# media per ogni aggettivo
mean_per_adj = df_bias.mean(axis=1)

# deviazione dalla media
df_bias = df_bias.sub(mean_per_adj, axis=0)

#display(df_bias)


# calcola il totale per aggettivo (somma sui dialetti + italiano)
df_heat = heatmap_df.copy()
df_heat["total"] = df_heat.sum(axis=1)

# calcola proporzione per ogni lingua
for col in ["dialetto", "italiano", "parmigiano", "siciliano"]:
    df_heat[col] = df_heat[col] / df_heat["total"]

#df_heat = df_heat.drop(columns="total")
#display(df_heat)

df_heat = df_heat.fillna(0)
pd.set_option("styler.render.max_rows", None)  # show all rows in Styler
pd.set_option("styler.render.max_columns", None)
import matplotlib

norm = matplotlib.colors.TwoSlopeNorm(vmin=0, vcenter=0.25, vmax=1)

data = df_heat.fillna(0)
gmap = data.apply(norm)

styled = (
    data.style
    .background_gradient(
        cmap="coolwarm",
        gmap=gmap,
        vmin=0,
        vmax=1,
        axis=None   #  THIS IS THE FIX
    )
)

styled.to_html("df_heat_full.html")
styled
```

```python


path_domanda_en="../test_multiple_napoli_question_english/all_answers_with_english_questions.jsonl"

df = pd.read_json(path_domanda_en, lines=True)


df['response'] = df['response_rephrased'].fillna(df['response'])
df["aggettivo"] = (
    df["prompt"]
    .str.strip()
    .str.rstrip(".")
    .str.split()
    .str[-1]
)
lista_aggettivi_original=["sporca","pulita","rumorosa","fredda","affettuosa","ritardataria","pia","onesta","furba","precisa","divertente","noiosa"]

filtered_agg=df[df['aggettivo'].isin(lista_aggettivi_original)]
filtered_language=filtered_agg[filtered_agg['language'].isin(['italiano','dialetto'])]
counts = filtered_language.groupby(["model_name", "language"]).size().reset_index(name="count")

df_yes_si = filtered_language[
    filtered_language["response"]
    .str.strip()
    .str.rstrip("!.,")
    .str.casefold()
    .isin(["yes", "si"])
]

grouped_by_model = df_yes_si.groupby(['aggettivo', 'language',"model_name"]).size().reset_index(name='count')



df_grouped_question_en=grouped_by_model.rename(columns={"count": "count_en"})

df_grouped_question_english=df_grouped_question_en.pivot_table(
    index=["aggettivo", "language"],
    columns="model_name",
    values="count_en",
    fill_value=0
)





df_heat = df_grouped_question_english.copy()

# total yes/si per (aggettivo, language)
df_heat["total"] = df_heat.sum(axis=1)

# divide each model by the row total
df_heat = df_heat.div(df_heat["total"], axis=0)

df_heat = df_heat.drop(columns="total").fillna(0)



import matplotlib

pd.set_option("styler.render.max_rows", None)
pd.set_option("styler.render.max_columns", None)

norm = matplotlib.colors.TwoSlopeNorm(vmin=0, vcenter=0.25, vmax=1)

data = df_heat
gmap = data.apply(norm)

styled = (
    data.style
    .background_gradient(
        cmap="coolwarm",
        gmap=gmap,
        vmin=0,
        vmax=1,
        axis=None
    )
)

styled.to_html("df_heat_models_aggettivo_language.html")
styled

#display(df_grouped_question_english)

#display(filtered_agg)

```
