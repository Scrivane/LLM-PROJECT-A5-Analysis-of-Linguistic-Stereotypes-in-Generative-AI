```python
import pandas as pd

fixed222_path_to_llm_log_final_analysis = "different_prompts_analisis.jsonl"
fixed222_path_to_llm_log_final_analysis = "final_global_log.jsonl"
fixed222_path_to_llm_log_final_analysis = "all_prompts_to_make.jsonl"


df = pd.read_json(fixed222_path_to_llm_log_final_analysis, lines=True)

# ==========================================================
# Generic helper to produce a table
# ==========================================================
def compute_counts(df, group_col, filter_col=None, filter_value=None, count_name=None):
    """
    df              dataframe
    group_col       column to group on (image_title or model_name)
    filter_col      optional column to filter (task1/task2)
    filter_value    value to filter on ("yes"/"no")
    count_name      name of the result column
    """
    
    # base table of unique values (ensures rows with zero counts appear)
    base = pd.DataFrame(df[group_col].unique(), columns=[group_col])
    
    # total tests per group
    total = df.groupby(group_col).size().reset_index(name='total_tests')

    # filtered subset
    if filter_col:
        df_filtered = df[df[filter_col] == filter_value]
        counts = df_filtered.groupby(group_col).size().reset_index(name=count_name)
    else:
        # no filtering requested
        counts = pd.DataFrame({group_col: [], count_name: []})

    # merge everything cleanly
    merged = (
        base
        .merge(counts, on=group_col, how='left')
        .merge(total, on=group_col, how='left')
        .fillna({count_name: 0})
    )

    merged[count_name] = merged[count_name].astype(int)



    return merged



def compute_counts2(
    df,
    group_cols,
    filter_col=None,
    filter_value=None,
    count_name=None,
    sum_cols=None,        # List of columns to sum over ALL rows in the group
    sort_by=None,
    ascending=None        # MUST be a list of bools if sort_by is a list
):
    """
    df : DataFrame
    group_cols : string or list of columns to group by
    filter_col : column to filter on before grouping (task1/task2)
    filter_value : value to filter on ("yes"/"no")
    count_name : name of the resulting count column (filtered)
    sum_cols : list of numeric columns to sum over all rows in the group
    sort_by : column or list of columns to sort by
    ascending : list of booleans (same length as sort_by)
    """

    # Ensure group_cols is a list
    if isinstance(group_cols, str):
        group_cols = [group_cols]

    # Ensure sort_by is a list
    if sort_by is not None and isinstance(sort_by, str):
        sort_by = [sort_by]

    # Validate ascending
    if sort_by is not None:
        if ascending is None:
            ascending = [True] * len(sort_by)
        elif not isinstance(ascending, list):
            raise ValueError("ascending must be a list of booleans")
        elif len(ascending) != len(sort_by):
            raise ValueError("ascending must have same length as sort_by")

    # Base table
    base = df[group_cols].drop_duplicates()

    # Total tests per group (UNFILTERED)
    total = df.groupby(group_cols).size().reset_index(name="total_tests")

    # Sum specified columns over ALL rows in the group
    if sum_cols:
        sum_df = df.groupby(group_cols)[sum_cols].mean().reset_index()
        sum_df[sum_cols] = sum_df[sum_cols].round(1)

    else:
        sum_df = pd.DataFrame(columns=group_cols + (sum_cols if sum_cols else []))


    # Initialize merged table
    merged = base.merge(total, on=group_cols, how="left").merge(sum_df, on=group_cols, how="left")

    # Filtered dataframe for counts
    if filter_col:
        df_filtered = df[df[filter_col] == filter_value]

        # Count passed tests
        if count_name:
            counts = (
                df_filtered.groupby(group_cols)
                .size()
                .reset_index(name=count_name)
            )
            merged = merged.merge(counts, on=group_cols, how="left")
            merged[count_name] = merged[count_name].fillna(0).astype(int)
    else:
        if count_name:
            merged[count_name] = 0

    # Percentage column
    if count_name:
        merged[f"{count_name}_pct"] = (
            merged[count_name] / merged["total_tests"] * 100
        ).round(2)

    # Sorting
    if sort_by is not None:
        merged = merged.sort_values(by=sort_by, ascending=ascending)
    



   
    return merged


# ==========================================================
# Generate all tables with one-liners
# ==========================================================
prompt_map = {
    """
this is the legend explaining what is represented in the ER diagram :
legend:
entities= rectangles
Relationship = rhombus
Attributes = pin attached to an entity or relationship
Identifier Attributes (Single): Represented by a black pin attached directly to an entity.
composite Identifier Attributes: Represented by a black pin attached to a curved line that connects to two or more white pins (the constituent attributes). This signifies that the attributes linked by the curved line together form the identifier.
Foreign Key= When a curved line originating from a black pin also connects to a relationship, it indicates that the identifier includes a foreign key
weak entity = entity that has a foreign key 
composite attributes = capsuled name attached to an entity , its subattributes are with pins linked to capsule .
attributes may have cardinality specified , if they have not , the default cardinality is (1,1)
recursive relationship= relationship between an entity and itself 
Generalization = directed arrow to the parent entity , originates from its children entities 
Generalization includes a role label telling if its partial or total and if it is overlapping or exclusive

Rules:
All entities should have at least one identifier attribute or an external identifier. (Except for child entities of a generalization which should not have an identifier attribute).
The weak entity must participates with cardinality (1,1) in his identifying relationship,
Attributes cardinalities can be omitted, if they are omitted they are considered with a cardinality of (1,1) .
Identifier attributes must have explicit or implicit cardinality of (1,1).
The parent composite attribute should have maximum cardinality equal to 1 .
All cardinalites of children attribute of a composite attribute are the same as their parent (even if they are not explicit). 
All participation of entities in a relationship must be expressed.
Relationship can have attributes associated. Relationship cant have identifier attributes.
In a ternary relationship all entities should participate with maximum cardinality equal to N .
In a generalization the type of  generalization must always be expressed .
On each side of the recursive relationship role name must be specified .


Task:
Are there any structural errors in this ER schema ?
if yes , explain the errors.
""": "legend+rules_prompt",
    """this is the legend explaining what is represented in the ER diagram :
legend:
entities= rectangles
Relationship = rhombus
Attributes = pin attached to an entity or relationship
Identifier Attributes (Single): Represented by a black pin attached directly to an entity.
composite Identifier Attributes: Represented by a black pin attached to a curved line that connects to two or more white pins (the constituent attributes). This signifies that the attributes linked by the curved line together form the identifier.
Foreign Key= When a curved line originating from a black pin also connects to a relationship, it indicates that the identifier includes a foreign key
weak entity = entity that has a foreign key 
composite attributes = capsuled name attached to an entity , its subattributes are with pins linked to capsule .
attributes may have cardinality specified , if they have not , the default cardinality is (1,1)

Generalization = directed arrow to the parent entity , originates from its children entities 
Generalization includes a role label telling if its partial or total and if it is overlapping or exclusive

Rules:
All entities should have at least one identifier attribute or an external identifier. (Except for child entities of a generalization which should not have an identifier attribute).
The weak entity must participates with cardinality (1,1) in his identifying relationship,
Attributes cardinalities can be omitted, if they are omitted they are considered with a cardinality of (1,1) .
Identifier attributes must have explicit or implicit cardinality of (1,1).
The parent composite attribute should have maximum cardinality equal to 1 .
All cardinalites of children attribute of a composite attribute are the same as their parent (even if they are not explicit). 
All participation of entities in a relationship must be expressed.
Relationship can have attributes associated. Relationship cant have identifier attributes.
In a ternary relationship all entities should participate with maximum cardinality equal to N .
In a generalization the type of  generalization must always be expressed .
On each side of the recursive relationship role name must be specified .

Task:
Are there any structural errors in this ER schema ?
if yes , explain the errors in a concise way. Do not add extra suggestion , just show the errors.
""": "legend+rules+concise_prompt",
   "\nthis is the legend explaining what is represented in th ER diagram :\nlegend:\nentities= rectangles\nRelationship = rhombus\nAttributes = pin attached to an entity or relationship\nIdentifier Attributes (Single): Represented by a black pin attached directly to an entity.\ncomposite Identifier Attributes: Represented by a black pin attached to a curved line that connects to two or more white pins (the constituent attributes). This signifies that the attributes linked by the curved line together form the identifier.\nForeign Key= When a curved line originating from a black pin  also connects to a relationship, it indicates that the identifier includes a foreign key\nweak entity = entity that has a foreign key \ncomposite attributes = white pin with a name attached to an entity , its subattributes are linked to it . Subattributes must have same cardinality of the composite attribute \nattributes may have cardinality specified , if they have not , the default cardinality is (1,1)\n\nGeneralization = directed arrow to the parent entity , originates from its children entities \nGeneralization includes a role label telling if it\u2019s partial or total and if it is overlapping or  exclusive\ntask:\nAre there any structural errors in this ER schema ?\nif yes , explain the errors\n\n"
   : "legend_prompt",
}


df["prompt"] = df["prompt"].map(prompt_map).fillna(df["prompt"])
df["answer_len"] = df["response"].str.len().fillna(0).astype(int)

pd.set_option("display.max_rows", None)
for i, row in df.iterrows():
    if (int(row['answer_len'])<100):
        print(f"Response: {row['response']}\nLength: {row['answer_len']}\n")



# Per image -------------------------------------
task1_image = compute_counts2(df, ["image_title"], "task1", "yes", "task1_yes_count")
task2_image = compute_counts2(df, ["image_title"], "task2", "no",  "task2_no_count")

# Per model -------------------------------------
task1_model = compute_counts2(df, ["model_name"], "task1", "yes", "task1_yes_count")
task2_model = compute_counts2(df, ["model_name"], "task2", "no",  "task2_no_count")

# ==========================================================
# Display results
# ==========================================================
print("Task1 == 'yes' per image")
display(task1_image)

print("\nTask2 == 'no' per image")
display(task2_image)

print("\nTask1 == 'yes' per model")
display(task1_model)

print("\nTask2 == 'no' per model")
display(task2_model)





task1_prompt_model = compute_counts2(df, ["prompt","model_name"], "task1", "yes", "task1_yes_count",sort_by=["model_name","task1_yes_count_pct"],ascending=[True,True],sum_cols=["answer_len"])
task2_prompt_model = compute_counts2(df, ["prompt","model_name"], "task2", "no",  "task2_no_count",sort_by=["model_name","task2_no_count_pct"],ascending=[True,True],sum_cols=["answer_len"])

# Per model -------------------------------------
task1_prompt_image = compute_counts2(df, ["image_title","model_name"], "task1", "yes", "task1_yes_count")
task2_prompt_image = compute_counts2(df, ["image_title","model_name"], "task2", "no",  "task2_no_count")
print("Task1 == 'yes' per image")
display(task1_prompt_model)

print("\nTask2 == 'no' per image")
display(task2_prompt_model)

print("\nTask1 == 'yes' per model")
display(task1_prompt_image)

print("\nTask2 == 'no' per model")
display(task2_prompt_image)



## image title prompt 

common_models_only=["qwen3-vl:235b-instruct-cloud","magistral-small-2509","meta-llama/llama-4-maverick-17b-128e-instruct","google/gemma-3-27b-it:free","gemini-2.5-flash"]
task1_prompt_image = compute_counts2(df, ["prompt","image_title"], "task1", "yes", "task1_yes_count",sort_by=["image_title","task1_yes_count_pct"],ascending=[True,True])


print("\nTask1 == 'yes' per model")
display(task1_prompt_image)


pd.set_option('display.max_colwidth', 70)  # or use a large number like 200
df_only_common_models = df[df["model_name"].isin(common_models_only)]
print("only common models used , task 1 ")
task1_image_common = compute_counts2(df_only_common_models, ["image_title"], "task1", "yes", "task1_yes_count",sort_by=['task1_yes_count_pct'],ascending=[True])
task2_image_common = compute_counts2(df_only_common_models, ["image_title"], "task2", "no",  "task2_no_count",sort_by=['task2_no_count_pct'],ascending=[True])
print("\nTask1 == 'yes' per model")
display(task1_image_common)

print("\nTask2 == 'no' per model , immagine pi missleading")

display(task2_image_common)

```
